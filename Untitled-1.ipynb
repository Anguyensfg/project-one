```python
df = pd.concat(
    map(pd.read_csv, ["raw_data/2017-18-vacancy.csv", "raw_data/2018-19-vacancy.csv", 
                        "raw_data/2019-20-vacancy.csv", "raw_data/2020-21-vacancy.csv"]), ignore_index=True)
df.shape
```

```python
Years = []
date_splited = []
dates = pd.Series(df['TENDate'])
for date in dates:
    if '/' in date:
        date_splited = date.split('/')
        Years.append(date_splited[2].strip())
    elif '-' in date:
        date_splited = date.split('-')
        Years.append(f'20{date_splited[1].strip()}')
    else:
        Years.append(date)

print(len(dates))
print(len(Years))
df['Year'] = Years
```
